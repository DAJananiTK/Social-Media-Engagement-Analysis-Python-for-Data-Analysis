```python
#social media engagement dataset
# Task 1: Data Import & Setup
# Import CSV using Pandas
# Check/convert data types
# Convert date columns to datetime

import pandas as pd

# Upload the CSV file
from google.colab import files
uploaded = files.upload()

# Load the dataset
df = pd.read_csv('social_media_engagement_5000.csv')

# Display the first 5 rows
print("First 5 Rows of the Dataset:")
display(df.head())


# Check data types
print("\nData Types:")
print(df.dtypes)

# Convert date column to datetime
# date column = 'posted_at'

df['posted_at'] = pd.to_datetime(df['posted_at'])

# Verify the conversion
print("\nUpdated Data Types:")
print(df.dtypes)

#  first five rows after conversion
display(df.head())
```

```python

# Task 2: Data Cleaning


# 2.1 cleaning the missing data

# Detect  missing values
print("Missing Values in Each Column:")
print(df.isnull().sum())

# Another way to detect missing values
print("\nUsing isna():")
print(df.isna().sum())


# Handle using dropna(), fillna(), median/mode, forward/backward-fill

# Select numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Fill numerical missing values with median
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())

# Select categorical columns
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

# Fill categorical missing values with mode
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Check missing values after cleaning
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

# 2.2 Duplicate Handling- Identify & remove duplicates

# Count duplicate rows
duplicate_count = df.duplicated().sum()
print("\nNumber of Duplicate Rows:", duplicate_count)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Verify duplicates are removed
print("Duplicate Rows After Removal:", df.duplicated().sum())

# Display dataset shape
print("\nDataset Shape After Cleaning:", df.shape)
```

```python
# 2.3 Data Formatting
# Fix incorrect data types
# Standardize categories (e.g., gender labels)
# Correct unrealistic values in likes, comments, shares


# Check the current data types
print("Data Types Before Formatting:\n")
print(df.dtypes)

# Fix Incorrect Data Types

# Convert age to integer
df['age'] = df['age'].astype(int)

# Convert likes, comments, and shares to integer
df['likes'] = df['likes'].astype(int)
df['comments'] = df['comments'].astype(int)
df['shares'] = df['shares'].astype(int)


# Display updated data types
print("\nData Types After Formatting:\n")
print(df.dtypes)

#Standardize categories (e.g., gender labels)

#  unique gender values before standardization
print("\nUnique Gender Values Before:")
print(df['gender'].unique())
# no standardization needed




# Correct unrealistic values in likes, comments, shares
# Check for negative values in likes, comments, and shares

print("Negative Likes:", (df['likes'] < 0).sum())
print("Negative Comments:", (df['comments'] < 0).sum())
print("Negative Shares:", (df['shares'] < 0).sum())
#all values are 0 , so need to correct the values



# 2.4  Feature Cleaning

# Extract Hashtag Count

# Create a new column to count the number of hashtags
df['hashtag_count'] = df['hashtags'].str.split().str.len()

# Display the first 5 rows
print("Hashtag Count:")
display(df[['hashtags', 'hashtag_count']].head())

# Clean Sentiment Labels


# Check unique sentiment values
print("\nUnique Sentiment Labels:")
print(df['sentiment'].unique())

# no need for cleaning the sentiment labels
#already cleaned and standardized
```

```python
#Task 3 — Data Exploration using Pandas
# Perform the following:
# View dataset structure using head(), tail(), shape, and columns


# Display first 5 rows
print("First 5 Rows:")
display(df.head())

# Display last 5 rows
print("\nLast 5 Rows:")
display(df.tail())

# Display dataset shape
print("\nDataset Shape:")
print(df.shape)

# Display column names
print("\nColumn Names:")
print(df.columns)



# Check data types and info with info() and dtypes

# Display dataset information
print("\nDataset Information:")
df.info()

# Display data types
print("\nData Types:")
print(df.dtypes)



# Generate summary statistics using describe()
# it is for numerical columns
print("\nSummary Statistics:")
display(df.describe())


# Analyze categorical distributions using value_counts(), unique(), and nunique().
# Gender Distribution
print("\nGender Distribution:")
print(df['gender'].value_counts())

# Post Type Distribution
print("\nPost Type Distribution:")
print(df['post_type'].value_counts())

# Sentiment Distribution
print("\nSentiment Distribution:")
print(df['sentiment'].value_counts())

# Unique values in Country
print("\nUnique Countries:")
print(df['country'].unique())

# Number of unique countries
print("\nNumber of Unique Countries:")
print(df['country'].nunique())



# Create a correlation matrix for numeric fields.

print("\nCorrelation Matrix:")
display(df.select_dtypes(include=['int64', 'float64']).corr())

# Use groupby() to summarize metrics (e.g., avg likes by post type, impressions by country).

print("\nAverage Likes by Post Type:")
display(df.groupby('post_type')['likes'].mean())

# Average Impressions by Country
print("\nAverage Impressions by Country:")
display(df.groupby('country')['impression_count'].mean())
```

```python
# Task 4 — Data Wrangling
# Use merge, concat, or join if combining DataFrames.(no merge or concat operations because of single dataset)
# Create new fields such as engagement_score, log-transformed metrics(optional), and hashtag count.
# Perform groupby summaries by post_type, country, and sentiment.


import numpy as np

#create new fields
# Create Engagement Score
df['engagement_score'] = df['likes'] + df['comments'] + df['shares']

# Create Log-Transformed Engagement Score
df['log_engagement_score'] = np.log1p(df['engagement_score'])

# Create Hashtag Count
df['hashtag_count'] = df['hashtags'].str.split().str.len()

# Display new columns
print("New Features:")
display(df[['engagement_score',
            'log_engagement_score',
            'hashtag_count']].head())


# GroupBy Analysis


# Group By Post Type
print("Average Likes by Post Type:")
display(df.groupby('post_type')['likes'].mean())

# Group By Country
print("\nAverage Impression Count by Country:")
display(df.groupby('country')['impression_count'].mean())

# Group By Sentiment
print("\nAverage Engagement Rate by Sentiment:")
display(df.groupby('sentiment')['engagement_rate'].mean())





```

```python
# Task 5 — Statistical Analysis
# Compute descriptive stats for “likes, comments, shares, watch_time,engagement_rate, followers” columns:

# Select numerical columns
stats_columns = ['likes', 'comments', 'shares',
                 'watch_time_sec', 'engagement_rate',
                 'follower_count']

#Mean, median, mode
print("Mean:")
print(df[stats_columns].mean())

print("\nMedian:")
print(df[stats_columns].median())


print("\nMode:")
display(df[stats_columns].mode())


# Standard deviation, variance
print("\nStandard Deviation:")
print(df[stats_columns].std())



print("\nVariance:")
print(df[stats_columns].var())


# Percentiles
print("\nPercentiles:")
display(df[stats_columns].quantile([0.25, 0.50, 0.75]))

# Skewness and kurtosis
print("\nSkewness:")
print(df[stats_columns].skew())


print("\nKurtosis:")
print(df[stats_columns].kurt())
```
